# merge the two tables and format for circos graph

```python
import pandas as pd
```

```python
# read both csv as dataaframe
counts = pd.read_csv('/RNA/VALENCIA/counts.csv')
counts=counts.rename(columns = {'Unnamed: 0':'geneID'})



pos= pd.read_csv('/RNA/VALENCIA/sorted', index_col = False, header = None, sep = '\t')
pos.columns = ['chr','geneID','start','end']
#pos.rename(columns={"0":"chr", "1":"geneID", "2":"start", "3":"end"})
```

```python
pos
```

# Merge the two dataframes based off of the column geneID

```python
# merge the data frames based on the gene ID
pos.merge(counts, on='geneID', how = 'left')

# perform merge and create a dict containing a dataframe for each sample
# create a list to iterate from
list = ['CLV.102.1.GHA' , 'CLV.102.2.GHA' , 'CLV.102.3.GHA' , 'CLV.102.1.PFA' , 'CLV.102.2.PFA' , 'CLV.102.3.PFA' ,'CLV.103.1.PFA','CLV.103.2.PFA','CLV.105.3.PFA','CLV.108.1.PFA','CLV.108.2.PFA','CLV.108.3.PFA','PV5.1.PFA','PV5.2.PFA','PV5.3.PFA']

# subset the dataframe into  individual dataframes and place inside a dict
samples = {}
for i in range(0,15):
    samples[i]= counts[['geneID',list[i]]]

# merge the dfs for each sample based on geneID    
df_dict = {}
for i in range(0,15):
    df_dict[i]=pos.merge(samples[i], on='geneID', how = 'left')
    #filter any value that doesn't match the left
    df_dict[i]=df_dict[i].dropna()
    #write the dfs to csv to be used in circos
    df_dict[i].to_csv('/RNA/circos/circos-0.69-9/hist2/hist'+str(i+1)+'.txt',sep='\t',header=False, index=False, columns=['chr','start','end',list[i]] )
```
